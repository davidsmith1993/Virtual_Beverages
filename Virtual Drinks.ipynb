# Virtual Beverage Flavor Sensations

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Import the dataset for the virtual beverage flavor sensations and convert to long form 

```python
"""import the overall dataset"""
dfall = pd.read_csv('C:/Users/Pablo/Desktop/Data Analysis - Vocktail and Voffee/Python/Vocktail Cleaned out Wide With composite Scores.csv') 


"""Import sensations, salty, sour, sweet, and bitter and then create dataframes for each"""

"""salty"""
dfsalt = dfall[['Sub ID', 'Color Salt', 'Salt and Sugar Salt', 'Smell Salt', 'Combination Salt']]
dfsaltlong = pd.melt(dfsalt, id_vars = ['Sub ID'], var_name = 'group', value_name = 'score')

"""sour"""
dfSour = dfall[['Sub ID', 'Color sour', 'Salt and Sugar sour', 'Smell sour', 'Combination sour']]
dfSourlong = pd.melt(dfSour, id_vars = ['Sub ID'], var_name = 'group', value_name = 'score')


"""sweet"""
dfSweet = dfall[['Sub ID', 'Color sweet', 'Salt and Sugar sweet', 'Smell sweet', 'Combination sweet']]
dfSweetlong = pd.melt(dfSweet, id_vars = ['Sub ID'], var_name = 'group', value_name = 'score')


"""bitter"""
dfBitter = dfall[['Sub ID', 'Color bitter', 'Salt and Sugar bitter', 'Smell bitter', 'Combination bitter']]
dfBitterlong = pd.melt(dfBitter, id_vars = ['Sub ID'], var_name = 'group', value_name = 'score')
```

```python
gr1 = dfsaltlong.groupby('group')['score'].aggregate(['mean','sem'])
gr2 = dfSourlong.groupby('group')['score'].aggregate(['mean','sem'])
gr3 = dfSweetlong.groupby('group')['score'].aggregate(['mean','sem'])
gr4 = dfBitterlong.groupby('group')['score'].aggregate(['mean','sem'])


gr1 = gr1.reindex(['Color Salt', 'Salt and Sugar Salt', 'Combination Salt', 'Smell Salt'])
gr2 = gr2.reindex(['Color sour', 'Salt and Sugar sour', 'Combination sour', 'Smell sour'])
gr3 = gr3.reindex(['Color sweet', 'Salt and Sugar sweet', 'Combination sweet', 'Smell sweet'])
gr4 = gr4.reindex(['Color bitter', 'Salt and Sugar bitter', 'Combination bitter', 'Smell bitter'])
```

Plot the flavor sensation ratings

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex=False, sharey=True, figsize=(15,7), dpi = 300)
plt.style.use(['default', 'seaborn-paper'])
ax1.bar(x=gr1.index, height=gr1['mean'], yerr=gr1['sem'], capsize=10, color = ('r','g','y', 'b'))
ax2.bar(x=gr2.index, height=gr2['mean'], yerr=gr2['sem'], capsize=10, color = ('r','g','y', 'b'))
ax3.bar(x=gr3.index, height=gr3['mean'], yerr=gr3['sem'], capsize=10, color = ('r','g','y', 'b'))
ax4.bar(x=gr4.index, height=gr4['mean'], yerr=gr4['sem'], capsize=10, color = ('r','g','y', 'b'))
ax3.set_ylabel('Score (Standard Error)', fontsize=12)
ax1.set_ylabel('Score (Standard Error)', fontsize=12)
ax1.set_xticklabels(['','','',''])
ax2.set_xticklabels(['','','',''])
ax3.set_xticklabels(['Color','Electric','Combination','Smell'])
ax4.set_xticklabels(['Color','Electric','Combination','Smell'])
ax1.set_title("Salt")
ax2.set_title("Sour")
ax3.set_title("Sweet")
ax4.set_title("Bitter")
ax1.set_ylim(1,4)
fig.suptitle('Flavor Sensation Ratings: Virtual Beverage', fontsize=16)
plt.savefig('C:/Users/Pablo/Desktop/Data Analysis - Vocktail and Voffee/Python/VirtualBeverageSensations.png')
plt.show()
```

# Virtual Beverage Liking Ratings

```python
"""Import liking, overall, taste, smell, and temp and then create dataframes for each"""

"""overall"""
dfOL = dfall[['Sub ID', 'Color OL', 'Salt and Sugar OL', 'Smell OL', 'Combination OL']]
dfOLlong = pd.melt(dfOL, id_vars = ['Sub ID'], var_name = 'group', value_name = 'score')

"""taste"""
dfTaste = dfall[['Sub ID', 'Color Taste', 'Salt and Sugar Taste', 'Smell Taste', 'Combination Taste']]
dfTastelong = pd.melt(dfTaste, id_vars = ['Sub ID'], var_name = 'group', value_name = 'score')


"""smell"""
dfSmell = dfall[['Sub ID', 'Color Smell Liking', 'Salt and Sugar Smell Liking', 'Smell Smell Liking', 'Combination Smell Liking']]
dfSmelllong = pd.melt(dfSmell, id_vars = ['Sub ID'], var_name = 'group', value_name = 'score')


"""temp"""
dfTemp = dfall[['Sub ID', 'Color Temp Like', 'Salt and Sour Temp Like', 'Smell Temp Like', 'Combination Temp Like']]
dfTemplong = pd.melt(dfTemp, id_vars = ['Sub ID'], var_name = 'group', value_name = 'score')


gr1 = dfOLlong.groupby('group')['score'].aggregate(['mean','sem'])
gr2 = dfTastelong.groupby('group')['score'].aggregate(['mean','sem'])
gr3 = dfSmelllong.groupby('group')['score'].aggregate(['mean','sem'])
gr4 = dfTemplong.groupby('group')['score'].aggregate(['mean','sem'])

gr1 = gr1.reindex(['Color OL', 'Salt and Sugar OL', 'Combination OL', 'Smell OL'])
gr2 = gr2.reindex(['Color Taste', 'Salt and Sugar Taste', 'Combination Taste', 'Smell Taste'])
gr3 = gr3.reindex(['Color Smell Liking', 'Salt and Sugar Smell Liking', 'Combination Smell Liking', 'Smell Smell Liking'])
gr4 = gr4.reindex(['Color Temp Like', 'Salt and Sour Temp Like', 'Combination Temp Like', 'Smell Temp Like'])
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, sharex=False, sharey=True, figsize=(15,7), dpi = 300)
ax1.bar(x=gr1.index, height=gr1['mean'], yerr=gr1['sem'], capsize=10, color = ('r','g','y', 'b'))
ax2.bar(x=gr2.index, height=gr2['mean'], yerr=gr2['sem'], capsize=10, color = ('r','g','y', 'b'))
ax3.bar(x=gr3.index, height=gr3['mean'], yerr=gr3['sem'], capsize=10, color = ('r','g','y', 'b'))
ax4.bar(x=gr4.index, height=gr4['mean'], yerr=gr4['sem'], capsize=10, color = ('r','g','y', 'b'))
ax3.set_ylabel('Score (Standard Error)', fontsize=12)
ax1.set_ylabel('Score (Standard Error)', fontsize=12)
ax1.set_xticklabels(['','','',''])
ax2.set_xticklabels(['','','',''])
ax3.set_xticklabels(['Color','Electric','Combination','Smell'])
ax4.set_xticklabels(['Color','Electric','Combination','Smell'])
ax1.set_title("Overall")
ax2.set_title("Taste")
ax3.set_title("Smell")
ax4.set_title("Temperature")
ax1.set_ylim(1,9)
fig.suptitle('Liking Ratings: Virtual Beverage', fontsize=16)
plt.savefig('C:/Users/Pablo/Desktop/Data Analysis - Vocktail and Voffee/Python/VirtualBeverageLiking.png')
plt.show()

```
